```python
#"Debt Collection"
#"Other financial service"
#"Credit reporting, credit repair services, or other personal consumer reports"
#"Mortgage"
#"Student loan"
#Bank Account- "Bank account or service";"Checking or savings account"
#"Money transfers"
#Credit Card- "Credit card or prepaid card"; "Credit card"; "Prepaid card"
#"Consumer Loan"
#Payday loans- "Payday loan, title loan, or personal loan"; "Payday loan";
#"Vehicle loan or lease"
#"Credit reporting"
#Virtual Currency- "Money transfer, virtual currency, or money service"; "Virtual currency"

# for i in range(len(complaints_df2['product'])):
#     if product == "Bank account or service":
#         produt.replace
#     #mask = df2['fyear']== 2000 + i
#     #year.append(2000+i)
#     #num.append(df2[mask]['cik'].nunique())   
    
# def myreplace(s):
#     for ch in ['%','$',',']:
#         s = s.replace(ch, '')
#     return s
# df['A'] = df['A'].map(myreplace)

# def product_bank_acc(x):
#     if x == "Bank account or service":
#         x = x.replace("Bank account or service" , "Bank Account") 
#     if x == "Checking or savings account":
#         x = x.replace("Checking or savings account", "Bank Account")
# complaints_df2['product'] = complaints_df2['product'].map(product_bank_acc)


def product_bank_acc(x):
    for i in ["Bank account or service", "Checking or savings account"]:
        x = x.replace(i,'Bank Account')
    return x
complaints_df2['product'] = complaints_df2['product'].map(product_bank_acc)

def credit_card(x):
    for i in ["Credit card or prepaid card", "Credit card", "Prepaid card"]:
        x = x.replace(i,'Credit Card')
    return x
complaints_df2['product'] = complaints_df2['product'].map(credit_card)

def payday_loans(x):
    for i in ["Payday loan, title loan, or personal loan", "Payday loan"]:
        x = x.replace(i,'Payday Loans')
    return x
complaints_df2['product'] = complaints_df2['product'].map(payday_loans)

def virtual_currency(x):
    for i in ["Money transfer, virtual currency, or money service", "Virtual currency"]:
        x = x.replace(i,'Virtual Currency')
    return x
complaints_df2['product'] = complaints_df2['product'].map(virtual_currency)
```
